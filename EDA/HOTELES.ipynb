```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python

df = pd.read_csv('/Users/felipeamezquita/Downloads/hotels.csv', encoding='ISO-8859-1')

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
# Mostrar los nombres de las columnas del DataFrame
print(df.columns)
```

```python

# Filtrar el DataFrame para mantener solo las filas donde countyName es "United States"
df_filtered = df.loc[df[' countyName'] == 'United States']

# Mostrar las primeras filas del DataFrame filtrado
df_filtered
```

```python

# Separar la columna 'cityName' en dos partes: Ciudad y Estado
# Utilizamos n=1 para limitar a una sola división
df_filtered[['City', 'State']] = df_filtered[' cityName'].str.split(',', n=1, expand=True)

# Limpiar los espacios en blanco en las nuevas columnas 'City' y 'State'
df_filtered['City'] = df_filtered['City'].str.strip()
df_filtered['State'] = df_filtered['State'].str.strip()

# Mostrar las primeras filas del DataFrame después de la separación y limpieza
print("\nPrimeras filas del DataFrame después de la separación:")
print(df_filtered[['City', 'State']].head())

# Mostrar los valores únicos en la columna 'State' para verificar el contenido
print("\nValores únicos en la columna 'State':")
print(df_filtered['State'].unique())
```

```python
df_filtered
```

```python
# Crear una lista con los nombres de los estados de la costa este
estados_costa_este = [
    'Maine', 'New Hampshire', 'Massachusetts', 'Rhode Island',
    'Connecticut', 'New York', 'New Jersey', 'Delaware',
    'Maryland', 'Virginia', 'North Carolina', 'South Carolina',
    'Georgia', 'Florida'
]

# Filtrar el DataFrame para mantener solo las filas con estados en la lista de la costa este
df_filtered = df_filtered[df_filtered['State'].isin(estados_costa_este)]

# Eliminar filas con valores nulos en la columna 'State' si es necesario
df_filtered = df_filtered.dropna(subset=['State'])

# Mostrar las primeras filas del DataFrame filtrado
print("\nPrimeras filas del DataFrame filtrado:")
print(df_filtered.head())
```

```python
df_filtered
```

```python

estado_codigo = {
    'Maine': 'ME',
    'New Hampshire': 'NH',
    'Massachusetts': 'MA',
    'Rhode Island': 'RI',
    'Connecticut': 'CT',
    'New York': 'NY',
    'New Jersey': 'NJ',
    'Delaware': 'DE',
    'Maryland': 'MD',
    'Virginia': 'VA',
    'North Carolina': 'NC',
    'South Carolina': 'SC',
    'Georgia': 'GA',
    'Florida': 'FL'
}

# Mapear el código del estado en una nueva columna 'StateCode'
df_filtered['StateCode'] = df_filtered['State'].map(estado_codigo)
```

```python

# Eliminar las columnas no deseadas
columns_to_remove = [' cityCode', ' Attractions', ' Description', ' FaxNumber', 
                      ' HotelFacilities', ' PhoneNumber', ' PinCode', ' HotelWebsiteUrl']
df_filtered = df_filtered.drop(columns=columns_to_remove, errors='ignore')

# Mostrar las primeras filas del DataFrame después de eliminar las columnas
df_filtered
```

```python

# Crear una columna con el formato "Ciudad, EstadoCódigo"
df_filtered['CityState'] = df_filtered['City'] + ', ' + df_filtered['StateCode']

# Contar el número de hoteles por ciudad
hotel_counts = df_filtered['CityState'].value_counts()

# Obtener el top 10 de ciudades con más hoteles
top_cities = hotel_counts.head(10)

# Obtener la suma total de hoteles por ciudad para el top 10
top_cities_sum = df_filtered[df_filtered['CityState'].isin(top_cities.index)] \
    .groupby('CityState').size()

# Imprimir el top 10 de ciudades con más hoteles
print("Top 10 ciudades con más hoteles:")
for city_state, count in top_cities.items():
    print(f"{city_state}: {count} hoteles")

# Visualizar los datos
plt.figure(figsize=(14, 6))

# Gráfico de la cantidad de hoteles por ciudad (Top 10)
plt.subplot(1, 1, 1)
sns.barplot(x=top_cities.index, y=top_cities.values, palette='viridis')

# Añadir etiquetas a las barras con el total de hoteles
for index, value in enumerate(top_cities):
    plt.text(index, value, f'{value}', ha='center', va='bottom')

plt.xticks(rotation=90)
plt.title('Top 10 Ciudades con Más Hoteles')
plt.xlabel('Ciudad, Código Estado')
plt.ylabel('Cantidad de Hoteles')

plt.tight_layout()
plt.show()
```

```python

# Encontrar el estado con más hoteles
state_counts = df_filtered['State'].value_counts()

# Gráfico de la cantidad de hoteles por estado
plt.subplot(1, 1, 1)
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Cantidad de Hoteles por Estado')
plt.xlabel('Estado')
plt.ylabel('Cantidad de Hoteles')

plt.tight_layout()
plt.show()
```
